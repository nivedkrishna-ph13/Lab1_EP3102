```python
import time
import csv
from datetime import datetime

name = input("Enter your name: ")
data_rows = [["Name", "Timestamp", "Time Delay ", "Number Entered"]]

last_time = 0
```

```python
while True:
    num = input("Enter a number between 0-9 (or 'quit' to exit): ")
    if num.lower() == "quit":
        break
    if num.isdigit() and 0 <= int(num) <= 9:
        now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        current_time = time.time()
        delay = "" if last_time is None else round(current_time - last_time, 2)
        last_time = current_time
        data_rows.append([name, now, delay, num])
    else:
        print("Invalid input. Try again.")
```

```python
with open("output.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data_rows)
```
